```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Binning 
`Binning` : suatu metode yang digunakan untuk mentransformasi data numerik menjadi kelompok-kelompok yang lebih kecil. Tujuannya untuk lebih mudah menganalisa kolom tersebut, selain itu bisa meningkatkan performa model(khususnya untuk model-model regresi)

```python
df=sns.load_dataset('tips')
df
```

```python
## buat histogram
sns.distplot(df['tip'],bins=5)
```

```python
bins=range(5,51,5)
for i in bins:
    print(f'bin {i}')
    sns.distplot(df['tip'],bins=i)
    plt.show()
```

## Binning : Equal Frequences

`Equal Frequences` : Tiap kelompok memiliki count yang sama
### pd.qcut

```python
df['bin_tip_eq_freq']=pd.qcut(df['tip'],5)
# df['bin_tip_eq_freq']=pd.qcut(df['tip'],5,labels=['very low','low','medium','high','very high'])
df
```

```python
df['bin_tip_eq_freq'].value_counts()
```

# soal 1 

coba lakukan binning untuk kolom total bill menggunakan pd.qcut(equal frequences), buat menjadi 8 kelompok, berikan labelnya dari 1,2,3,4,...8

```python
df['bin_totalbill_eq_freq']=pd.qcut(df['total_bill'],8,labels=np.arange(1,9))
df
```

## Binning : Equal Interval

### pd.cut

```python
df['bin_tip_eq_int']=pd.cut(df['tip'],5)
# df['bin_tip_eq_int']=pd.cut(df['tip'],5,labels=['very low','low','medium','high','very high'])
df
```

## Univariate Analysis

```python
#equal frequence
df['bin_tip_eq_freq'].value_counts().reset_index()
```

```python
sns.countplot(x=df['bin_tip_eq_freq'])
```

```python
#equal interval
df['bin_tip_eq_int'].value_counts().reset_index()
```

```python
sns.countplot(x=df['bin_tip_eq_int'])
```

## Bivariate Analysis
tip vs total_bill

```python
sns.heatmap(df[['size','tip','total_bill']].corr(method='spearman'),annot=True,cmap='viridis')
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip')
sns.regplot(data=df,x='total_bill',y='tip',scatter=False, color='black',ci=None)
plt.title('Total Bill vs Tip')
```

```python
df_agg=df.groupby('bin_totalbill_eq_freq')['tip'].mean().reset_index()
df_agg
```

```python
sns.barplot(data=df_agg,x='bin_totalbill_eq_freq',y='tip')
plt.xlabel('bin_totalbill')
plt.ylabel('Average Tip')
plt.title('Total Bill vs Tip')
```

## Soal 2 (2 poin)

Buat analisis bivariat untuk melihat hubungan antara umur dengan survived (titanic seaborn), gunakan 5 bin (equal frequence)

```python
df=sns.load_dataset('titanic')
df['age']=round(df['age'].fillna(df['age'].mean())).astype(int)
df['bin_age']=pd.qcut(df['age'],5)
df
```

```python
df[df['age'].isnull()]
```

```python
df_agg=df.groupby('bin_age')['survived'].mean().reset_index()
df_agg
```

```python
sns.barplot(data=df_agg,x='bin_age',y='survived')
plt.xlabel('bin_age')
plt.ylabel('Survived Rate')
plt.title('Total Bill vs Tip')
```

## Multivariate Analysis

`tip vs total_bill (sex)`

```python
df=sns.load_dataset('tips')
df['bin_totalbill_eq_freq']=pd.qcut(df['total_bill'],5)
# df['bin_tip_eq_freq']=pd.qcut(df['tip'],5,labels=['very low','low','medium','high','very high'])
df
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex')
sns.regplot(data=df[df['sex']=='Female'],x='total_bill',y='tip',scatter=False,ci=None)
sns.regplot(data=df[df['sex']=='Male'],x='total_bill',y='tip',scatter=False,ci=None)
plt.title('Total Bill vs Tip')
```

```python
df_agg=df.groupby(['bin_totalbill_eq_freq','sex'])['tip'].mean().reset_index()
```

```python
sns.barplot(data=df_agg,x='bin_totalbill_eq_freq',y='tip',hue='sex')
plt.xlabel('bin_totalbill')
plt.ylabel('Average Tip')
plt.xticks(rotation=45)
plt.title('Total Bill vs Tip')
```
